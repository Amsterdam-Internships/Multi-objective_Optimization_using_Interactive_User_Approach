```python
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
```

```python
data = pd.read_csv('output-EI_2-20.csv')
```

```python
num_queries = data['Number of Queries']
regret = data['Regret']
```

```python
scale = MinMaxScaler()
regret_normalized = scale.fit_transform(regret.values.reshape(-1,1))
```

```python
plt.plot(regret, num_queries)
plt.xlabel('Regret')
plt.ylabel('Number of Queries')
plt.title('Regret vs Number of Queries')
plt.show()
```
