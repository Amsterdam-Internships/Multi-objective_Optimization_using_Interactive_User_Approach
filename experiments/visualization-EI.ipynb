```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import csv
```

```python
data = pd.read_csv('regret-csvs/regret-BLR_2-100.csv')
```

```python
steps = data['Step']
regret = data['Regret']
```

```python
# regret = [float(value[1:-1]) for value in regret]
# regret_vals = np.array([regret[0] for regret in regret])
# regret_round = np.round(regret, 3)
```

```python
df = pd.DataFrame({'Step': steps, 'Regret': regret})
sns.histplot(data=df, x='Step', bins=50, stat='count', alpha=0.8)
# stat is the aggregate statistic to compute in each bin. Stat is chosen to be count to show the 
# number of observations in each bin
plt.xlabel('Number of Query')
plt.ylabel('Number of Observations in each bin')
plt.title('Distribution of Regret for each query')
plt.show()
```

```python
df['Average Regret'] = df['Regret'].apply(lambda x: float(x.strip('[]')))
```

```python
avge = df.groupby(steps)['Average Regret'].mean()
plt.plot(avge.index, avge.values)

plt.xlabel('Number of Queries')
plt.ylabel('Average Regret')
plt.title('Average Regret as a function of the Number of Queries')
plt.show()
```
