```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
data = pd.read_csv('output-EI_2-20.csv')
```

```python
num_queries = data['Number of Queries']
regret = data['Regret']
comp_time = data['Computation Time']
```

```python
regret = [float(value[1:-1]) for value in regret]
regret_round = np.round(regret, 3)
```

```python
# sns.histplot(num_queries, label='Number of Queries', color='b')
# sns.histplot(regret, label='regret', stat='density', color='g' )
# plt.barh(regret, num_queries)
plt.bar(num_queries, regret_round)
plt.ylabel('Regret')
plt.xlabel('Number of Queries')
plt.title('Number of Queries vs Regret with 2 objectives')
plt.grid(True)
plt.show()
```

```python
data['Regret'] = data['Regret'].apply(lambda x: float(x[1:-1]))
avg_regret = data.groupby('Number of Queries')['Regret'].mean()
plt.bar(avg_regret.index, avg_regret)

plt.xlabel('Number of Queries')
plt.ylabel('Avg Regret')
plt.title('Avg regret per Number of Queries')
# plt.legend()
plt.show()
```

```python
# data['Regret'] = data['Regret'].apply(lambda x: float(x[1:-1]))
sns.displot(regret_round, bins=30, kde=True)
plt.xlabel('Num Queries')
plt.ylabel('Regret')
plt.title('Distribution of Regret by Num Queries')
plt.show()
```

```python
hist_bins = 10

data['Number of Queries'] = pd.cut(data['Number of Queries'], bins=hist_bins)

sns.histplot(data=data, x='Regret', bins='auto', kde=True)
# sns.histplot(data[data['Number of Queries']==0]['Regret'], label='Number of Queries =1', kde=True)

plt.xlabel('Regret')
plt.ylabel('Density')
plt.title('Distribution of Regret for different number of queries')
plt.legend()
plt.show()
```
