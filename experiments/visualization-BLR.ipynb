```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import csv
```

```python
data = pd.read_csv('output-EI_5-100.csv')
```

```python
num_queries = data['Number of Queries']
regret = data['Regret']
```

```python
# regret = [float(value[1:-1]) for value in regret]
# regret_vals = np.array([regret[0] for regret in regret])
# regret_round = np.round(regret, 3)
```

```python
df = pd.DataFrame({'Number of Queries': num_queries, 'Regret': regret})
```

```python
sns.histplot(data=df, x='Number of Queries', bins=20, stat='count')
# stat is the aggregate statistic to compute in each bin. Stat is chosen to be count to show the 
# number of observations in each bin
plt.xlabel('Number of Queries')
plt.ylabel('Number of Observations in each bin')
plt.title('Distribution of Number of Queries as a function of the Regret')

plt.xlim(-10, 290)
plt.ylim(0, 40)
plt.show()
```

```python
df['Average Regret'] = df['Regret'].apply(lambda x: float(x.strip('[]')))
df['Average Regret'] = df['Average Regret'].round(2)
df
```
